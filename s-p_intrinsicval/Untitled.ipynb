```python
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36 OPR/118.0.0.0"
}

all_data = []

for ticker in tickers:
    print(f"Veri Ã§ekiliyor: {ticker}")
    url = f"https://finance.yahoo.com/quote/{ticker}/key-statistics?p={ticker}"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    rows = soup.find_all("tr", class_="yf-kbx2lo")
    
    data = {"Symbol": ticker}
    for row in rows:
        cells = row.find_all("td")
        if len(cells) >= 2:
            key = cells[0].text.strip()
            val = cells[1].text.strip()
            data[key] = val
    
    all_data.append(data)
    time.sleep(2)  


df = pd.DataFrame(all_data)


df = df.sort_values("Symbol").reset_index(drop=True)


print(df.head())
```

```python
from scipy.stats import percentileofscore

cols = df.columns[1:6]

for col in cols:
    new_col = col + "_percentile"
    df[new_col] = df[col].apply(lambda x: 1 - percentileofscore(df[col], x) / 100)
```

```python
percentile_cols = [col + "_percentile" for col in df.columns[1:6]]

df["Total_Percentile_Score"] = df[percentile_cols].sum(axis=1)
```
